```python
import networkx as nx
import numpy as np
import tensorflow as tf
```

```python
def to_edgelist(file):
    mylines = []
    #get rid of num of neighbors and node tag
    with open(file ,'r') as p:
        for line in p:                   
            mylines.append(line[4:])
    #add node to start of each line    
    for i in range(len(mylines)):
        mylines[i] = f'{i} ' + mylines[i]
    #rewrite file so it is in edgelist format
    with open(file,'w') as p:
        p.writelines(mylines)
    #make the graph
    G = nx.read_adjlist(file)
    #clear list just in case
    mylines.clear()
    return G
```

```python
list_of_graphs=[]
with open('600EnzymeGraphs.txt','r') as main:
    #populate lines in main file as a list
    lines = main.readlines()
    
    #initilize start and end for for loop
    start = 1
    num_nodes = (int(lines[0][0:3])) + 1
    
    #intialize *.txt name
    num_nodes_name = int(lines[0][0:3])
    nodes_class = int(lines[0][3:4])
    with open('Enzyme_labels.txt','w') as t:
        t.write(f'{nodes_class}' + '\n')
    t.close()
        
    x=0
    #make x the number of graphs being worked with      
    while x < 600:
        #open the files for the graph
        with open('graph_'+f'{x}'+'_'+ f'{num_nodes_name}'+'_'+f'{nodes_class}'+'.txt','w') as t:
            
            #Write each block from main file into their own file
            for line in lines[start:num_nodes]:
                t.write(f'{line}') 
            #close the graph so we can open it up again
            t.close()
            #convert the graph that we just wrote into edgelist format using method from above
            #can use to_edglist to turn into a graph and add it to a list then this function can return the list
            G = to_edgelist('graph_'+f'{x}'+'_'+ f'{num_nodes_name}'+'_'+f'{nodes_class}'+'.txt')
            t.close()
            list_of_graphs.append(G)
            #update start and end postions, while loop, and file name
            try: 
                num_nodes_name = int(lines[num_nodes][0:-2])
                nodes_class = int(lines[num_nodes][-2])
                with open('Enzyme_labels.txt','a') as t:
                    t.write(f'{nodes_class}' + '\n')
                t.close()
            except IndexError:
                break
            x = x + 1
            start = num_nodes + 1
            num_nodes = (int(lines[num_nodes][0:-2])) + 1 + num_nodes
            #will produce error at end because cannot see when the end of the main file is
            
```

```python
def feature_to_mean(dictionary):
    feature_list = []
    feature_list = list(dictionary.values())
    mean_feature_list = np.mean(feature_list)
    return mean_feature_list
```

```python
with open('Enzymes_features.txt','w') as t:
    for i in range(len(list_of_graphs)):
    
        dc_dict = nx.degree_centrality(list_of_graphs[i])
        cc_dict = nx.closeness_centrality(list_of_graphs[i])
        bc_dict = nx.betweenness_centrality(list_of_graphs[i])
        hc_dict = nx.harmonic_centrality(list_of_graphs[i])
        #ec_dict = nx.eigenvector_centrality(list_of_graphs[i])
        clust_dict = nx.clustering(list_of_graphs[i])
        #pr_dict = nx.pagerank(h)
        
        t.write(f'{feature_to_mean(dc_dict)}' + f'{feature_to_mean(cc_dict)}' + f'{feature_to_mean(bc_dict)}' + f'{feature_to_mean(hc_dict)}' + f'{feature_to_mean(clust_dict)}' + '\n' )
    
```
